```python
import gzip
import pandas as pd
import numpy as np
```

```python
print("Extracting Patient Labels...")
labels = []
# 1. Read just the text header to find the disease states
with gzip.open("./GSE13159_series_matrix.txt.gz", "rt") as f:
    for line in f:
        # Relaxed search condition to simply look for our target keywords in the characteristics row
        if line.startswith("!Sample_characteristics_ch1") and ("leukemia" in line.lower() or "healthy" in line.lower()):
            parts = line.strip().split('\t')[1:]
            labels = [p.replace('"', '').lower() for p in parts]
            break
print(f"Successfully extracted {len(labels)} labels!")
```

```python
print("Loading Expression Matrix (This takes ~30 seconds)...")
# 2. Read the table directly, automatically skipping all metadata rows starting with '!'
expression_data = pd.read_csv("./GSE13159_series_matrix.txt.gz", sep="\t", comment="!", index_col=0)

# 3. Transpose so patients are rows and genes are columns
df = expression_data.T
df['Target_Label'] = labels

print("Filtering for the 'Big Four' Leukemias + Healthy...")
conditions = [
    df['Target_Label'].str.contains('healthy'),
    df['Target_Label'].str.contains('acute lymphoblastic'),
    df['Target_Label'].str.contains('acute myeloid'),
    df['Target_Label'].str.contains('chronic lymphocytic'),
    df['Target_Label'].str.contains('chronic myelogenous') | df['Target_Label'].str.contains('chronic myeloid')
]
```

```python
# Map to integers: 0=Healthy, 1=ALL, 2=AML, 3=CLL, 4=CML
choices = [0, 1, 2, 3, 4]
df['Target'] = np.select(conditions, choices, default=-1)

# Drop everything that isn't one of our 5 classes
clean_df = df[df['Target'] != -1].copy()
clean_df = clean_df.drop(columns=['Target_Label'])

print(f"Success! Final Dataset Shape: {clean_df.shape[0]} patients, {clean_df.shape[1]-1} genes.")
print("Class breakdown:")
print(clean_df['Target'].value_counts())

clean_df.to_csv("Leukemia_5Class_Dataset.csv", index=False)
print("Saved as Leukemia_5Class_Dataset.csv. Ready for ElasticNet.")
```
